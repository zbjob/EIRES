```python

import matplotlib.pyplot as plt
import pandas as pd
import matplotlib 
import itertools
from matplotlib.ticker import MaxNLocator

data = pd.read_csv("pixel.csv", delimiter=',', header=0) 
Woolsey_precisions =[]
Woolsey_recalls =[]
plt.figure(figsize=(8.5,4.5))

time = list(map(int, set(data["Time"])))
camp_precisions = []
camp_recalls = []
county_precisions = []
county_recalls = []
kincade_precisions = []
kincade_recalls = []
for index, row in data.iterrows():
    row["F1"] = float(row["F1"])
    if row["Dataset"] == 'Woolsey':
        Woolsey_precisions.append(row["F1"]/100)
        Woolsey_recalls.append(row["Recall"]/100)
    if row["Dataset"] == 'Camp':
        camp_precisions.append(row["F1"]/100)
        camp_recalls.append(row["Recall"]/100)
    if row["Dataset"] == 'County':
        county_precisions.append(row["F1"]/100)
        county_recalls.append(row["Recall"]/100)
    if row["Dataset"] == 'Kincade':
        kincade_precisions.append(row["F1"]/100)
        kincade_recalls.append(row["Recall"]/100)
plt.figure(1)
ax1 = plt.subplot(111)
plt.grid(True, color='#8b8b8d', linestyle='dotted', alpha=0.7)
ax1.plot(time, county_precisions, '-.', marker='2', ms=7, label='County',linewidth=1.5)
ax1.plot(time, camp_precisions, ':', marker='o', ms=7, label='Camp',linewidth=1.5)
ax1.plot(time, Woolsey_precisions, '-', marker='s',ms=7, label='Woolsey', linewidth=1.5)
ax1.plot(time, kincade_precisions, '--', marker='d', ms=7, label='Kincade',linewidth=1.5)
# ax1.set_xlabel('Time')
ax1.set_ylabel('Weighted F1', fontsize=18)
ax1.set_ylim(0.5, 1)
plt.locator_params(axis='y', nbins=6)
ax1.xaxis.set_major_locator(MaxNLocator(integer=True))
ax1.tick_params(axis="x", labelsize=14)
ax1.tick_params(axis="y", labelsize=14)
ax1.set_xlabel('Time', fontsize=18)
#ax1.xaxis.grid(False)
# ax1.title.set_text('Precision')
# ax1.set_title('Precision', x=0.88, y=0.1, fontsize=12,fontweight='bold')
handles, labels = ax1.get_legend_handles_labels()
plt.legend(handles[::1], labels[::1],bbox_to_anchor=(0., 1.02, 2., .102), loc=3,
           ncol=4, frameon=False, fontsize=14)

# ax2 = plt.subplot(212)
# plt.grid(True, color='#8b8b8d', linestyle='dotted', alpha=0.5)
# ax2.plot(time,camp_recalls, ':', marker='o', ms=7, label='Camp',   linewidth=1.5)
# ax2.plot(time, Woolsey_recalls, '-', marker='s',ms=7, label='Woolsey',  linewidth=1.5)
# ax2.plot(time, county_recalls, '-.', marker='2', ms=7,  label='Taboose',linewidth=1.5)
# ax2.plot(time, kincade_recalls, '--', marker='d', ms=7, label='Walker',  linewidth=1.5)
# ax2.set_ylim(0.5, 1)
# # ax2.set_title('Recall', x=0.85, y=0.1, fontsize=12,fontweight='bold')
# ax2.set_xlabel('Time', fontsize=18)
# ax2.set_ylabel('Recall', fontsize=18)
# plt.locator_params(axis='y', nbins=6)
# ax2.tick_params(axis="x", labelsize=14)
# ax2.tick_params(axis="y", labelsize=14)
# ax2.xaxis.set_major_locator(MaxNLocator(integer=True))
#plt.show()
plt.savefig('../../figures/pixel_acc.png')

```
